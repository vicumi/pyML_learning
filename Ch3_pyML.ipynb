## First steps with scikit-learn

Loading the Iris data fron scikit-learn. Here, the third column (number 2) represents the petal length, and the fourth column (number 3) the petal width of the flower samples. The classes are already converted to integer labels where 0=Iris-Setosa, 1=Iris-Versicolor, 2=Iris-Virginica.

```python
from sklearn import datasets
import numpy as np

iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target

print("Class labels: ",np.unique(y))
```

Splitting data into 70% and 30% test data with function `train_test_split()`.  
Note: `train_test_split()` has already randomized  the data before splitting in case that training dataset contains only tag 0 and tag 1 and testing dataset contains only 2.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
# a little test
print("Labels counts in y: ", np.bincount(y))
print("Labels counts in y_train: ", np.bincount(y_train))
print("Labels counts in y_teat: ", np.bincount(y_test))
```

### Standardizing the features:

`sc.fit()` estimates the μ and σ.  
`sc.transform` standardizes the features data.  
$x_{std} = \frac{x-μ}{σ}$

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

## Training a perceptron via scikit-learn

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)
```

```python
y_pred = ppn.predict(X_test_std)
print("Misclassified samples: {}".
      format((y_test != y_pred).sum()))
```

Envaluing the training result (accuracy):
* using `metrics` module in sklearn
* uing the envaluating method of the classifier Perceptron (named ppn in this sample)

```python
from sklearn.metrics import accuracy_score

print("Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred)))
```

```python
print("Accuracy: {:.2f}".format(ppn.score(X_test_std, y_test)))
```

## Ploting the decision regions

Redefining the plot_decision_region function from chapter 2. The test data are highlighted with circled points.

```python
%matplotlib inline
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator and color map
    markers = ('s','x','o','^','v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:,0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:,1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl,0],
                    y=X[y == cl,1],
                    alpha=0.8,
                    c=colors[idx],
                    marker=markers[idx],
                    label=cl,
                    edgecolor='black')
    
    # hightlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolors='black',
                    alpha=1.0,
                    linewidths=1,
                    marker='o',
                    s=100,
                    label='test set')
```

Training a percrptron model using the standardized training data:

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined_std = np.hstack((y_train, y_test))

plot_decision_regions(X=X_combined_std,
                      y=y_combined_std,
                      classifier=ppn,
                      test_idx=range(105,150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='best')

plt.tight_layout()

plt.show()
```

As is dipicted in the graph above, those three kinds of Iris are not able to classified completely with linear decision regions.  
In fact, the perceptron algorithm never converges to incompletely linear sparable data sets.

# Modeling class probabilities via logistic regression 

<br/>

## Logistic regression intuition and conditional probabilities

The most important function is *sigmod function* :
$$\phi(z) = \frac{1}{1+e^{-z}}$$
where z is ‘net input’, which is a linear combination of weights and sample features, $z = \boldsymbol{w}^{T}\boldsymbol{x} = w_0x_0+w_1x_1+...+w_mx_m$.

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmod(z):
    return 1.0 / (1.0 + np.exp(-z))

z = np.arange(-7, 7, 0.1)
phi_z = sigmod(z)

plt.plot(z,phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')

# y axis ticks and gridline
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)

plt.tight_layout()
plt.show()
```

## Learing the weights of the logistic cost function

The simplified cost function：
$$
J(\boldsymbol{w})=J(\phi(z),y;\boldsymbol{w})=\begin{cases}
-log(\phi(z))& \text{if y=1}\\
-log(1-\phi(z))& \text{if y=0}
\end{cases}
$$

```python
def cost_1(z):
    return - np.log(sigmod(z))

def cost_0(z):
    return - np.log(1-sigmod(z))

z = np.arange(-10, 10, 0.1)
phi_z = sigmod(z)

c1 = [cost_1(x) for x in z]
plt.plot(phi_z, c1, label='J(w) if y=1')

c0 = [cost_0(x) for x in z]
plt.plot(phi_z, c0, linestyle='--', label='J(w) if y=0')

plt.ylim(0.0, 5.1)
plt.xlim([0,1])
plt.xlabel('$\phi(z)$')
plt.ylabel('J(w)')
plt.legend(loc='best')
plt.tight_layout()

plt.show()
```

It can be observed that the cost is approaching 0 if the data is predicted as 'kind 1' correctly (solid line), vice versa (dotted line).

## Trackling overfitting via regularization

An image explaining overfitting:
<img src="./images/03_overfitting.png" width=700 />

<br/>
L2 norm:
$$\frac{\lambda}{2} \left\| \boldsymbol{w} \right\|^{2}
=\frac{\lambda}{2} \sum_{j=1}^{m}w_{j}^{2}$$
where $\lambda$ is called regularization parameter。  

It is **in inverse propotion to the coefficient 'C'** in class `LogisticRegression`. (We will go further on this coefficient in the following part of SVM.)

For more information about regularization, see here: [What is regularization?](https://charlesliuyx.github.io/2017/10/03/%E3%80%90%E7%9B%B4%E8%A7%82%E8%AF%A6%E8%A7%A3%E3%80%91%E4%BB%80%E4%B9%88%E6%98%AF%E6%AD%A3%E5%88%99%E5%8C%96/)

```python
weights, params = [], []
for c in np.arange(-5,5):
    lr = LogisticRegression(C=10.**c, random_state=1,
                            solver='liblinear', 
                            multi_class='auto')
    lr.fit(X_train_std, y_train)
    weights.append(lr.coef_[1])
    params.append(10.**c)
    
weights = np.array(weights)
plt.plot(params, weights[:, 0], label='petal length')
plt.plot(params, weights[:, 1], linestyle='--', label='petal width')
plt.ylabel('weight coefficient')
plt.xlabel('C')
plt.legend(loc='best')
plt.xscale('log')

plt.show()
```

# Maximum margin classification with support vector machines (SVM)

SVM is designed to maximize the margin:
<img src="./images/03_SVM0.png" width=700 />

## Maximum margin intuition
Hyperplane: 
$$\boldsymbol{w}^{T}\boldsymbol{x}+b=0$$
Margin:
$$\frac{\boldsymbol{w}^{T}(\boldsymbol{x_+}-\boldsymbol{x_-})}{\left\|\boldsymbol{w}\right\|}=\frac{2}{\left\|\boldsymbol{w}\right\|}$$

## Dealing with the nonlinearly seperable case using slack variables

Minimizing the object function:
$$\frac{1}{2}{\left\|\boldsymbol{w}\right\|}^2+C(\sum_i \xi^{(i)})$$
Control the punishment of classification error with the variable $C$. The bigger is $C$, the larger is the punishment and more strict is in classification. 

<img src="./images/03_SVM1.png" width=600 />

```python
from sklearn.svm import SVC

svm = SVC(kernel="linear", C=1.0, random_state=1)
svm.fit(X_train_std, y_train)

plot_decision_regions(X_combined_std,
                      y_combined_std,
                      classifier=svm,
                      test_idx=range(102,150))
plt.xlabel("petal length [std]")
plt.ylabel("petal width [std]")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

### Using the kernel trick to find separating hyperplanes in higher dimensional space

Definition: Kernel Function
$$K(\boldsymbol{x}^{(i)},\boldsymbol{x}^{(j)})
=\phi(\boldsymbol{x}^{(i)})^T\phi(\boldsymbol{x}^{(j)})$$

One of the most popular function is RBF:
$$K(\boldsymbol{x}^{(i)},\boldsymbol{x}^{(j)})
=exp(-  \frac
        {\left\|
          \boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}
        \right\|^2}
        {2\sigma^2}
     )$$
Also:
$$K(\boldsymbol{x}^{(i)},\boldsymbol{x}^{(j)})
=exp(-  \gamma
        \left\|
          \boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}
        \right\| ^2
     )$$
where $\gamma=\frac{1}{2\sigma^2}$

Construct an XOR data set with random noise:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)
X_xor = np.random.randn(200,2)
y_xor = np.logical_xor(X_xor[:, 0]>0, X_xor[:, 1]>0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1,1],
            c='b', marker='*', label='1')
plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1,1],
            c='r', marker='s', label='-abs1')

plt.xlim([-3,3])
plt.ylim([-3,3])
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=1, gamma=0.10, C=10.0)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)

plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

## Decision tree learning

### Maxmizing information gain - getting the  most bang for buck
Definition: Information Gain
$$IG(D_p,f)=I(D_p)-\sum_{j=1}^{m}\frac{N_j}{N_p}I(D_j)$$
where:  
&emsp; $D_p$: parant node  
&emsp; $D_j$: No.j child node  
&emsp; $I$: impurity content  
&emsp; $N_p$: the number of samples in parent node  
&emsp; $N_j$: the number of samples in No.j child node  

In fact, most software libraries only implements the binary decision tree, which means each parent node only prossesses two child nodes $D_{left}$ and $D_{right}$.    
<br/>
Three criterion:
* Entropy:
$$I_H(t)=-\sum_{i=1}^{c}p(i|t)log_2p(i|t)$$
* Gini impurity:
$$I_G(t)=1-\sum_{i=1}^{c}p(i|t)^2$$
* Misclassificatioin error:
$$I_E(t)=1-max\{p(i|t)\}$$

where $p(i|t)$ is probability of a node $t$ belonging to a class $c$ sample.

```python
import matplotlib.pyplot as plt
import numpy as np

def gini(p):
    return p * (1 - p) + (1 - p) * (1 - (1 - p))

def entropy(p):
    return - p * np.log2(p) - (1 - p) * np.log2((1-p))

def error(p):
    return 1 - np.max([p,1-p])

x = np.arange(0.0, 1.0, 0.01)

ent = [entropy(p) if p != 0 else None for p in x]
sc_ent = [e * 0.5 if e else None for e in ent]
err = [error(i) for i in x]

fig = plt.figure
ax = plt.subplot(111)
for i, lab, ls, c, in list(zip([ent, sc_ent, gini(x), err],
                           ['Entropy', 'Entropy(scaled)', 'Gini Impurity', 'Misclassification Error'],
                           ['-', '-', '--', '-.'],
                           ['black', 'lightgray', 'red', 'green', 'cyan'])):
    line = ax.plot(x, i, label=lab, linestyle=ls, lw=2, color=c)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),
              ncol=5, fancybox=True, shadow=False)
ax.axhline(y=0.5, linewidth=1, color='k', linestyle='--')
ax.axhline(y=1.0, linewidth=1, color='k', linestyle='--')
plt.ylim([0, 1.1])
plt.xlabel('p(i=1)')
plt.ylabel('Impurity Index')
plt.show()
```

Minkowski distance：
$$d(\boldsymbol{x}^{(i)},\boldsymbol{x}^{(j)})
=\sqrt[p]{\sum_k {\left| x_k^{(i)}-x_k^{(j)} \right|}^p}$$

When:  
$p=2$, it becomes Euclidean distance;  
$p=1$, it becomes Manhattan distance.
